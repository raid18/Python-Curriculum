#The volume of a sphere with radius r is 4/3 π r^3. What is the volume of a sphere with radius 5?

```python
import math 
n = int(input("Enter Radius: "))                           #import always in the begining !!
def volume_sphere(r):                                      
  
  volume = 4/3 * (math.pi) * r**3

  return volume
print("the volume is",volume_sphere(n))                   #if I want to pass the variable I have to call the function 

```

Write a Program to convert time given in seconds to hours, minutes and seconds. 

```python
sec = 27006
hh = sec//(60*60)
remaining_seconds = sec%(60*60)
#print(remaining_seconds)                #that is not 'miniutes' this is 'remaining minutes'        
                                        #Floor/integer divison gives you the rounded off value to the lower number
mm = remaining_seconds//60 
remaining_sec = sec%60
print(remaining_sec,"seconds")   
print(mm,"minute")    
print(hh, "hours")

```
